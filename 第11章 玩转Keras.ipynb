```python
# 导入必要的库和模块
import tensorflow as tf  # TensorFlow 是一个开源的机器学习框架，提供了高级API，如Keras，用于构建和训练深度学习模型
from tensorflow.keras.datasets import cifar10  # cifar10 是TensorFlow Keras中提供的CIFAR-10数据集模块
from tensorflow.keras.models import Sequential  # Sequential 是Keras提供的顺序模型，用于按顺序堆叠神经网络层
from tensorflow.keras.layers import Dense, Dropout, Flatten  # 导入全连接层(Dense)、Dropout层和扁平化层(Flatten)
from tensorflow.keras.utils import to_categorical  # to_categorical 用于将标签转换为one-hot编码形式

# 加载 CIFAR-10 数据集并进行预处理
(x_train, y_train), (x_test, y_test) = cifar10.load_data()  # 加载CIFAR-10数据集，返回训练集和测试集
x_train = x_train.astype('float32') / 255  # 将训练集的像素值归一化到0-1之间
x_test = x_test.astype('float32') / 255  # 将测试集的像素值归一化到0-1之间
y_train = to_categorical(y_train, num_classes=10)  # 将训练集标签转换为one-hot编码形式
y_test = to_categorical(y_test, num_classes=10)  # 将测试集标签转换为one-hot编码形式

# 创建模型
model = Sequential()  # 创建一个顺序模型

# 添加第一个全连接层，输入形状为图片扁平化后的大小
model.add(Flatten(input_shape=x_train.shape[1:]))  # 将输入图像扁平化
model.add(Dense(512, activation='relu'))  # 添加一个具有512个节点的全连接层，激活函数为ReLU
model.add(Dropout(0.5))  # 添加Dropout层，丢弃率为0.5，以防止过拟合

# 添加第二个全连接层
model.add(Dense(256, activation='relu'))  # 添加一个具有256个节点的全连接层，激活函数为ReLU
model.add(Dropout(0.3))  # 添加Dropout层，丢弃率为0.3

# 添加第三个全连接层
model.add(Dense(128, activation='relu'))  # 添加一个具有128个节点的全连接层，激活函数为ReLU
model.add(Dropout(0.2))  # 添加Dropout层，丢弃率为0.2

# 添加输出层
model.add(Dense(10, activation='softmax'))  # 输出层有10个节点，对应CIFAR-10数据集的10个类别，激活函数为Softmax

# 编译模型
model.compile(loss='categorical_crossentropy',  # 损失函数为交叉熵损失，适用于多分类问题
              optimizer='adam',  # 优化器为Adam
              metrics=['accuracy'])  # 评估指标为准确率

# 训练模型
model.fit(x_train, y_train,  # 训练数据和标签
          batch_size=128,  # 每批次处理128个样本
          epochs=10,  # 训练10轮
          validation_data=(x_test, y_test),  # 验证数据和标签
          shuffle=True)  # 每个epoch前打乱训练数据

# 评估模型
scores = model.evaluate(x_test, y_test, verbose=1)  # 评估模型在测试集上的表现，verbose=1表示显示进度条
print('Test loss:', scores[0])  # 打印测试集上的损失
print('Test accuracy:', scores[1])  # 打印测试集上的准确率
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import matplotlib.pyplot as plt

# 生成随机数据集
np.random.seed(0)  # 设置随机种子，确保每次运行生成相同的数据
X_train = np.random.rand(100, 10)  # 生成100个样本，每个样本有10个特征
y_train = np.random.randint(2, size=(100, 1))  # 生成100个二分类标签

# 定义不同的初始学习率
lrs = [0.001, 0.01, 0.1]

# 训练多个模型并记录结果
histories = []
for lr in lrs:
    # 创建模型
    model = Sequential()
    model.add(Dense(64, activation='relu', input_dim=10))  # 第一层，64个节点，ReLU激活函数
    model.add(Dense(1, activation='sigmoid'))  # 输出层，1个节点，Sigmoid激活函数

    # 编译模型
    optimizer = Adam(learning_rate=lr)  # 使用Adam优化器，设置学习率为lr
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])  # 编译模型，损失函数为二分类交叉熵，评估指标为准确率

    # 训练模型
    history = model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=0)  # 训练模型，50个周期，每批次32个样本，不打印训练过程
    histories.append(history)  # 记录每个模型的训练历史
```

```python
# 可视化结果
plt.figure(figsize=(12, 3))
for i, history in enumerate(histories):
    plt.subplot(1, len(histories), i+1)
    plt.plot(history.history['loss'], label='lr={}'.format(lrs[i]))
    plt.title('Training Loss')
    plt.xlabel('Epoch')
    plt.legend()
plt.show()

plt.figure(figsize=(12, 3))
for i, history in enumerate(histories):
    plt.subplot(1, len(histories), i+1)
    plt.plot(history.history['accuracy'], label='lr={}'.format(lrs[i]))
    plt.title('Training Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
plt.show()
```

```python
import numpy as np  # 导入NumPy库，用于数值计算
import matplotlib.pyplot as plt  # 导入Matplotlib库，用于绘图
from keras.datasets import mnist  # 从Keras中导入MNIST数据集模块

# 加载MNIST数据集
(x_train, y_train), (x_test, y_test) = mnist.load_data()  # 加载MNIST数据集，返回训练集和测试集

# 对训练集和测试集进行预处理
x_train = x_train.astype('float32') / 255.  # 将训练集的像素值归一化到0-1之间
x_test = x_test.astype('float32') / 255.  # 将测试集的像素值归一化到0-1之间
```

```python
from keras.models import Sequential  # 导入Keras的顺序模型类
from keras.layers import Dense  # 导入Keras中的全连接层

def create_model(num_neurons):
    """
    创建一个简单的全连接神经网络模型。
    
    参数:
    num_neurons (int): 第一个全连接层中的神经元数量。
    
    返回:
    model (Sequential): 编译后的Keras顺序模型。
    """
    model = Sequential()  # 创建一个顺序模型

    # 添加输入层和隐藏层
    model.add(Dense(num_neurons, activation='relu', input_shape=(784,)))  # 
    # 添加一个全连接层，包含num_neurons个神经元，激活函数为ReLU
    # input_shape参数指定了输入数据的形状，MNIST图像展平后的形状为784（28x28像素）

    # 添加输出层
    model.add(Dense(10, activation='softmax'))  #
    # 添加一个全连接层，包含10个神经元（对应MNIST数据集中的10个类别）
    # 激活函数为Softmax，用于输出概率分布

    # 编译模型
    model.compile(optimizer='adam',  # 使用Adam优化器
                  loss='sparse_categorical_crossentropy',  # 使用稀疏分类交叉熵作为损失函数
                  metrics=['accuracy'])  # 监测准确率作为评估指标

    return model  # 返回编译后的模型
```

```python
def train_and_evaluate_model(model, x_train, y_train, x_test, y_test, num_epochs=10):
    """
    训练并评估模型。
    
    参数:
    model (Sequential): 待训练的Keras模型。
    x_train (numpy.ndarray): 训练集的输入数据。
    y_train (numpy.ndarray): 训练集的目标数据。
    x_test (numpy.ndarray): 测试集的输入数据。
    y_test (numpy.ndarray): 测试集的目标数据。
    num_epochs (int): 训练的轮数，默认为10。
    
    返回:
    accuracy (float): 测试集上的准确率。
    train_loss (list): 每个epoch的训练损失。
    val_loss (list): 每个epoch的验证损失。
    """
    # 训练模型
    history = model.fit(x_train, y_train,  # 训练数据和目标
                        epochs=num_epochs,  # 训练轮数
                        validation_data=(x_test, y_test),  # 验证数据和目标
                        verbose=0)  # 不打印训练过程信息

    # 评估模型
    _, accuracy = model.evaluate(x_test, y_test, verbose=0)  # 评估模型在测试集上的表现，verbose=0表示不打印评估过程信息

    # 返回测试集上的准确率以及训练和验证损失的历史记录
    return accuracy, history.history['loss'], history.history['val_loss']
```

```python
def main():
    # 定义不同的神经元数量
    neuron_counts = [16, 32, 64, 128, 256]
    accuracies = []  # 存储每个模型的准确率
    losses = []  # 存储每个模型的训练损失
    val_losses = []  # 存储每个模型的验证损失

    # 对于每种神经元数量，创建模型并训练
    for num_neurons in neuron_counts:
        model = create_model(num_neurons)  # 创建模型
        # 训练模型，并获取准确率、训练损失和验证损失
        accuracy, loss, val_loss = train_and_evaluate_model(
            model, 
            x_train.reshape(-1, 784),  # 将输入数据展平成784维
            y_train, 
            x_test.reshape(-1, 784),  # 将输入数据展平成784维
            y_test
        )
        accuracies.append(accuracy)  # 记录准确率
        losses.append(loss)  # 记录训练损失
        val_losses.append(val_loss)  # 记录验证损失

    # 可视化结果
    plt.figure(figsize=(12, 6))
    plt.subplot(121)  # 创建左图
    plt.plot(neuron_counts, accuracies, 'o-')  # 绘制准确率与神经元数量的关系图
    plt.xlabel('Number of neurons')  # 设置x轴标签
    plt.ylabel('Accuracy')  # 设置y轴标签
    plt.title('Model accuracy vs number of neurons')  # 设置图标题

    plt.subplot(122)  # 创建右图
    for i, num_neurons in enumerate(neuron_counts):
        plt.plot(np.arange(1, 11), losses[i], label=f'{num_neurons} neurons')  # 绘制每个模型的训练损失曲线
    plt.xlabel('Epochs')  # 设置x轴标签
    plt.ylabel('Training loss')  # 设置y轴标签
    plt.title('Training loss vs number of epochs')  # 设置图标题
    plt.legend()  # 显示图例

    plt.tight_layout()  # 自动调整子图参数，使之填充整个图像区域
    plt.show()  # 显示图像

```

```python
main()
```

```python
# 加载MNIST数据集
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# 对训练集图像数据进行预处理
train_images = train_images.reshape((60000, 28 * 28))  # 将训练集图像数据展平为一维向量，形状变为 (60000, 784)
train_images = train_images.astype('float32') / 255  # 将训练集图像数据类型转换为浮点数，并将像素值归一化到0-1之间

# 对测试集图像数据进行预处理
test_images = test_images.reshape((10000, 28 * 28))  # 将测试集图像数据展平为一维向量，形状变为 (10000, 784)
test_images = test_images.astype('float32') / 255  # 将测试集图像数据类型转换为浮点数，并将像素值归一化到0-1之间
```

```python
from keras.models import Sequential  # 导入Keras的顺序模型类
from keras.layers import Dense  # 导入Keras中的全连接层
from keras import regularizers  # 导入Keras中的正则化器

# 创建一个顺序模型
model = Sequential([
    # 第一个全连接层
    Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),  
    # 包含128个神经元，激活函数为ReLU
    # 使用L2正则化，正则化系数为0.001，以减少过拟合

    # 第二个全连接层
    Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),  
    # 同样包含128个神经元，激活函数为ReLU
    # 使用相同的L2正则化

    # 输出层
    Dense(10, activation='softmax')  
    # 包含10个神经元，激活函数为Softmax，用于输出概率分布
])
```

```python
# 编译模型
model.compile(optimizer='adam',  # 使用Adam优化器
              loss='sparse_categorical_crossentropy',  # 使用稀疏分类交叉熵作为损失函数
              metrics=['accuracy'])  # 监测准确率作为评估指标

# 训练模型
history = model.fit(train_images, train_labels,  # 训练数据和标签
                    epochs=20,  # 训练20轮
                    batch_size=64,  # 每个批次处理64个样本
                    validation_data=(test_images, test_labels))  # 使用测试集作为验证集
```

```python
from keras.callbacks import TensorBoard
# 创建TensorBoard回调
tensorboard_callback = TensorBoard(log_dir='./logs', histogram_freq=1)

# 训练模型
history = model.fit(train_images, train_labels,  # 训练数据和标签
                    epochs=20,  # 训练20轮
                    batch_size=64,  # 每个批次处理64个样本
                    validation_data=(test_images, test_labels),  # 使用测试集作为验证集
                    callbacks=[tensorboard_callback])  # 使用TensorBoard回调记录训练过程
```
